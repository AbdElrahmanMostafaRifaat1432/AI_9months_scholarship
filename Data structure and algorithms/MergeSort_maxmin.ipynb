```python
def merge_sort(arr):
    
    # first we will partition the data untill we reach 1 element
    if len(arr)==1:
        return arr
    
    elif len(arr) > 1:
        mid = len(arr)//2
        left = arr[:mid]
        right = arr[mid:]
        
        # then we will use recurrence to partition
        merge_sort(left)
        merge_sort(right)
        
        leftptr = 0
        rightptr = 0
        ptr = 0
        # then will compare the 2 sorted arrays
        while leftptr < len(left) and rightptr < len(right):
            if left[leftptr] < right[rightptr]:
                arr[ptr] = left[leftptr]
                leftptr+=1
                
            else:
                arr[ptr] = right[rightptr]
                rightptr+=1
                
            ptr+=1
            
            
        # there can be left or right that is still not concatenated
        
        while leftptr < len(left):
            arr[ptr] = left[leftptr]
            leftptr += 1
            ptr += 1

        while rightptr < len(right):
            arr[ptr] = right[rightptr]
            rightptr += 1
            ptr += 1
```

```python
arr = [1, 8, 4, 10, 2, 6]
merge_sort(arr)
print("Sorted array:", arr)
```

# T(n) = 2T(n/2)+n
# = O(nlogn) using master limit theorem

```python
def max_min(arr1, low, high):
    
    # this  is the base case either one element or two element
    if low == high:
        return arr1[low], arr1[low]
    
    # always return the maximum to the left
    elif high == low + 1:
        if arr1[low] < arr1[high]:
            return arr1[high], arr1[low]
        else:
            return arr1[low], arr1[high]

    # if I have more than 2 elements then I will partition
    else:  
        mid = (low + high) // 2

        # Recursively find max and min in the left and right halves
        leftMax , leftMin = max_min(arr1, low, mid)
        rightMax , rightMin = max_min(arr1, mid + 1, high)

        # Compare max and min of two halves
        last_max = max(leftMax, rightMax)
        last_min = min(leftMin, rightMin)

        return last_max, last_min
```

```python
# Example usage:
arr = [1, 8, 4, 10, 2, 6]
mostmax, mostmin = max_min(arr, 0, len(arr) - 1)
print("maximum is:", mostmax)
print("minimum is:", mostmin)
```
