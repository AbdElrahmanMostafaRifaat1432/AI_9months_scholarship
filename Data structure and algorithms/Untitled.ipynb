```python
class Spanning_tree:
    def __init__(self,numofnodes):
        self.num = numofnodes
        self.adjacent = [[]for i in range(numofnodes)]
        
    def edge(self , n1 ,n2 ):
        self.adjacent[n1].append(n2)
    
    def depth_first(self):
        # initiallize visited with false
        visited = [False for i in range(self.num)]
        stack = []
        stack.append(0)
        
        while(len(stack)):
            s = stack[-1] # get the last element in stack
            stack.pop() # then pop the last element
            
            if(not visited[s]): # if node is not visited 
                print(s , " ")
                visited[s] = True
                
            # get all the children
            
            i =len(self.adjacent[s])
            while i > 0 :
                i-=1
                if (not visited[self.adjacent[s][i]]):
                    stack.append(self.adjacent[s][i])
    
    def breadth_first(self):
        # initiallize visited with false
        visited = [False for i in range(self.num)]
        queue = []
        queue.append(0)
        
        while(len(queue)):
            q = queue[0] # get the first element in queue
            queue.pop(0) # then pop the first element
            
            if(not visited[q]): # if node is not visited 
                print(q , " ")
                visited[q] = True
                
            # get all the children
            
            for n in self.adjacent[q] :
                if (not visited[n]):
                    queue.append(n)
    

                
            
        
        
```

```python
sptree = Spanning_tree(7)
sptree.edge(0,1)
sptree.edge(0,2)
sptree.edge(1,3)
sptree.edge(1,4)
sptree.edge(2,5)
sptree.edge(2,6)
```
