```python
def outliers(name):
    q1 = df[name].quantile(0.25)
    q3 = df[name].quantile(0.75)
    iqr = q3 - q1
    print(iqr)
    upper = q3 + 1.5*iqr
    lower = q1 - 1.5*iqr
    return df[name][~((df[name]>upper) |(df[name]<lower))]
```

```python
outliers('sepal_length')
```
