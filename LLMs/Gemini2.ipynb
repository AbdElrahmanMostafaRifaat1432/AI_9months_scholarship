```python
def getSentiment(message , profileName):
     
    response = model.generate_content(f"""
        \nI have differnt users comments regarding fashion from different posts in an instagram fashion profile called {profileName}  , the comments can be in any language.          
        \n I want to classify the sentiment of each comment into either positve or neutral or negative (Don't add new sentiment other than those three)
        \n if a comment is showing interest about this fashion or asking about it's details to buy it classify it to positive
        \n if a comment is not related to the fashion classify it as neutral
        \n Just classify the sentiment one time and give me the output nothing else.                                                            
        \nI want the output to be in the following format:
        comment | Sentiment of this comment
        \n Don't create any other response beside the above output please
        \n classify all comments don't leave any comment without classification to either positive or neutral or negative
        \n here are 80 comments at maximum seperated by '\n' {message}                                                                         
    """)
    #print(response.text)
    return response.text
```
