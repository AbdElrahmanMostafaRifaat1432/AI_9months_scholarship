```python
import re

lines = response.text.strip().split('\n')[3:]

# Initialize a list to hold the parsed data
data = []

# Define a pattern to capture the postid, caption, and fashion items correctly
pattern = re.compile(r'(\d+)\s*\|\s*(\[.*\])')

for line in lines:
    match = pattern.match(line.strip())
    if match:
        postid, fashion_items = match.groups()
        data.append([postid, fashion_items])

# Create a DataFrame
df3 = pd.DataFrame(data, columns=['postid', 'fashion items'])

# Convert 'postid' to int64
df3['postid'] = df3['postid'].astype('int64')

# Convert 'fashion items' from string to list
# df3['fashion items'] = df3['fashion items'].apply(lambda x: eval(x))

def convert_to_list(item):
    # Replace outer single quotes with double quotes
    item = item.replace("'", '"')
    return eval(item)

df3['fashion items'] = df3['fashion items'].apply(convert_to_list)
df3.rename(columns={'postid': 'post_id'}, inplace=True)
```
