```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
for i in range(10):
    x = 3 * np.arange(0, 5, 0.01)
    y = 5 * np.sin(x)
    y += np.random.normal(0, 1, len(x))
    v = y[0]
    beta = 0.1 * i
    beta_bias = 0.3
    fitted_y = np.zeros_like(y)
    for j in range(len(y)):
        v = beta * v + (1 - beta) * y[j]
        v_bias = v / (1 - beta_bias**(i + 1))
        fitted_y[j] = v_bias
        
    plt.figure(i + 1)
    plt.figure(figsize=(20, 8))
    plt.suptitle(f'Beta = {beta}')

    plt.subplot(1, 3, 1)
    sns.scatterplot(x=x, y=y)

    plt.subplot(1, 3, 2)
    plt.plot(x, y)

    plt.subplot(1, 3, 3)
    sns.scatterplot(x=x, y=y)
    plt.plot(x, fitted_y, color='red')

    plt.show()
```
