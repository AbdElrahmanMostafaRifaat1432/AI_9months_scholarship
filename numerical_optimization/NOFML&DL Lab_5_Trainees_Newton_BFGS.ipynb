- Draw the functions and check out the minimum.

- The stop condition is grad < epsilon.
- Try different epsilons and initial conditions.
- Compare the results with Gradient descent.
##### The function should return the solution and the gradient.

##### Notes:
- You will need to pass the expression of the function, first, and second derivatives to the optimizer.
- You can use lambda expression to construct the function, first, and second derivatives.

```python
x = np.arange(-20,35)
f = lambda x: x**2 -15*x + 1
plt.plot(x , f(x))
plt.show()

x = np.arange(-20,35)
f_first = lambda x: 2*x -15
plt.plot(x , f_first(x))
plt.show()


x = np.arange(-20,35)
f_second = lambda x: 2+0*x
plt.plot(x , f_second(x))
plt.show()
```

```python
def gradient_descent(x0 , eps ,lr , f_first ):
    i=0
    while True:
        
        if (np.abs(f_first(x0))<eps):
                break;
        x0 = x0 - lr*f_first(x0)
        i+=1
        
    
    print( "Found solution after ", i , "iterations")
    print("x_min = ", x0)
    print( "Gradient = ", np.abs(f_first(x0)))
    
    
    
```

```python
gradient_descent(40,0.1,0.01 , f_first)
```

```python
# Gradient descent with learning rate = 0.01, x0=40, epsilon = 0.1
```

```python
def Newtons_no_learning(x0 , eps , f_first , f_second ):
    
    x0 = x0 - (f_first(x0)/f_second(x0))
    
    print( "Found solution after ", 1 , "iterations")
    print("x_min = ", x0)
    print( "Gradient = ", np.abs(f_first(x0)))
    
    
    
```

```python
Newtons_no_learning(40 , 0.1 , f_first , f_second)
```

```python
# Newton's method without learning rate, x0=40, epsilon = 0.1
```

```python
def Newtons_yes_learning(x0 , eps ,lr , f_first , f_second  ):
    i=0
    while True:
        if (np.abs(f_first(x0))<eps):
                break;
        x0 = x0 - lr*(f_first(x0)/f_second(x0))
        i+=1
        
    print( "Found solution after ", i , "iterations")
    print("x_min = ", x0)
    print( "Gradient = ", np.abs(f_first(x0)))
```

```python
Newtons_yes_learning(40,0.1,0.8, f_first , f_second)
```

```python
def gradient_descent_multi(x0 , eps ,lr , f_firstXy  ):
    i=0
    while True:
        if (np.linalg.norm(f_firstXy(x0[0],x0[1]))<eps):
                break;
        x0 = x0 - lr*f_firstXy(x0[0],x0[1])
        i+=1
        
    
    print( "Found solution after ", i , "iterations")
    print("x_min = ", x0)
    print( "Gradient = ", f_firstXy(x0[0],x0[1]))
```
