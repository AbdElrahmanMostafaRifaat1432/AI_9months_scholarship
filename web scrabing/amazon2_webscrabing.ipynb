```python
from selenium import webdriver
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.amazon.in/gp/bestsellers/books/")
```

```python
blocks = driver.find_elements("xpath" ,"//div[@id=\"gridItemRoot\"]")  
```

```python
import pandas as pd
data = { "title":[] ,"price":[] }
df  = pd.DataFrame.from_dict(data)
df
```

```python
i = 0
for block in blocks:
    #name = block.find_element("xpath",".//span[@class=\"a-size-base-plus a-color-base a-text-normal\"]")
    name = block.find_element("xpath",".//div[@class=\"_cDEzb_p13n-sc-css-line-clamp-1_1Fn1y\"]")

    print("name = " , name.text)
    try:
        price = block.find_element("xpath",".//div[@class=\"a-row\"]")
        print("price = " , price.text)
    except :
        price = None
        print("price = ")

    df.loc[i] = [name.text , price.text]
    i+=1
        

        
    
```
