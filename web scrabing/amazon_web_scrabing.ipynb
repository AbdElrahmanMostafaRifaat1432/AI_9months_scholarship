```python
from selenium import webdriver
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.amazon.eg/s?bbn=21832883031&rh=n%3A21832883031%2Cp_n_availability%3A21909182031&dc&language=en&qid=1703171305&rnid=21909180031&ref=lp_21832883031_nr_p_n_availability_1")
```

```python
blocks = driver.find_elements("xpath" , "//div[@class=\"a-section a-spacing-base\"]")
```

```python
i = 0
for block in blocks:
    name = block.find_element("xpath",".//span[@class=\"a-size-base-plus a-color-base a-text-normal\"]")
    print("name = " , name.text)
    
    try:
        price = block.find_element("xpath",".//span[@class=\"a-price-whole\"]")
        print("price = " , price.text)
    except :
        print("price = ")
        
    numvotes = block.find_element("xpath",".//a[@class=\"a-link-normal s-underline-text s-underline-link-text s-link-style\"]//span[@class=\"a-size-base s-underline-text\"]")
    print("number of votes = " , numvotes.text )
    
    
    rating = block.find_element("xpath",".//div[@class=\"a-row a-size-small\"]//span")
    r = rating.get_attribute('aria-label')
    print("rating is : " , r)
    
    urls = block.find_element("xpath",".//img[@class = \"s-image\"]")
    src= urls.get_attribute('src')
    print("urls : " , src)
    
    images = block.find_element("xpath",".//img[@class = \"s-image\"]")
    src= urls.get_attribute('src')
    
    response = requests.get(src)
    img = Image.open(BytesIO(response.content))
    plt.imshow(img)
    plt.show()
    img.save("img"+str(i)+".jpg")
    i+=1
    
    
    
    
    print("##################################################################################################################")
```
