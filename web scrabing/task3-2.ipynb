```python
from selenium import webdriver
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.boxofficemojo.com/chart/top_lifetime_gross/?area=XWW")
```

```python
blocks = driver.find_elements("xpath" ,"//tbody/tr")  
```

```python
import pandas as pd
data = { "title":[] ,"lifeTime":[] , "Releasedate":[] }
df  = pd.DataFrame.from_dict(data)
df
```

```python
i = 0
for block in blocks:
    if i!= 0 :
        title = block.find_element("xpath",".//td[@class=\"a-text-left mojo-field-type-title\"]/a ")
        print("name = " , title.text)
        

        lifeTime = block.find_element("xpath",".//td[@class=\"a-text-right mojo-field-type-money\"]")
        print("lifeTime = " , lifeTime.text)
        
        Releasedate = block.find_element("xpath",".//td[@class=\"a-text-left mojo-field-type-year\"]/a")
        print("Releasedate = " , Releasedate.text)
        
        df.loc[i] = [title.text , lifeTime.text ,Releasedate.text ]
    i+=1
        

        
    
```
